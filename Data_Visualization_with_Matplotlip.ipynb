```python
import matplotlib.pyplot as plt
```

```python
x=[1, 2, 3, 4, 5]
y=[2, 3, 5, 7, 11]  
plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.title('Sample Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)  
```

```python
x=[1, 2, 3, 4, 5]
y=[2, 3, 5, 7, 11]

## Create a customized line plot

plt.plot(x, y, color='red')
plt.title('Customized Line Plot')
```

```python
## Multiple Plots
## Sample data

x=[1, 2, 3, 4, 5]
y1=[2, 3, 5, 7, 11]
y2=[1, 4, 6, 8, 10]
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 2)
plt.plot(x, y1, marker='o', color='blue')
plt.title('First Plot')
plt.xlabel('X-axis')
plt.ylabel('Y1-axis')
plt.grid(True)

plt.subplot(1, 2, 1)
plt.plot(x, y2, marker='x', color='green')
plt.title('Second Plot')
plt.xlabel('X-axis')
plt.ylabel('Y2-axis')
plt.grid(True)
```

```python
### Bar plot

categories = ['A', 'B', 'C', 'D', 'E']  # Fixed the categories list to have 5 elements
vales = [5, 7, 3, 8, 6]
## Bar Plot
plt.bar(categories, vales, color='purple')
```

```python
## Histograms

data= [1, 2, 2, 3, 3, 3, 4, 5, 5, 6, 7, 8, 9]

##Create a histogram

plt.hist(data, bins=5, color='orange')
```

```python
## create a scatter plot
# sample data
x=[ 1, 2, 3, 4, 5]
y=[2, 3, 5, 7, 11]
plt.scatter(x, y, color='red', marker='x')
# Show all plots
```

```python
## Create a Pi Chart

labels= ['Category A', 'Category B', 'Category C']
sizes= [15, 30, 45]
explode=(0.2,0,0,)
plt.pie(sizes, labels=labels,explode=explode,autopct='%1.1f%%', startangle=140,shadow=True)
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is a circle.

plt.show()  # Display all plots
```
