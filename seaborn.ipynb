```python
import seaborn as sns
```

```python
## Basic Seaborn Plotting Example

tips=sns.load_dataset('tips')
tips
```

```python
import matplotlib.pyplot as plt
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='day')
plt.title('Scatter Plot of Total Bill vs Tip')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.show()
```

```python
##Line plot
sns.lineplot(x='size', y='total_bill', data=tips, hue='day')
plt.title('Line Plot of Total Bill by Size')
plt.xlabel('Size of Party')
plt.ylabel('Total Bill ($)')
plt.show()
```

```python
## Category Plot
sns.barplot(x='day', y='total_bill', data=tips)
plt.title('Box Plot of Total Bill by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill ($)')
plt.show()
```

```python
##Box Plot
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Box Plot of Total Bill by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill ($)')
plt.show()
```

```python
## Violin Plot
sns.violinplot(x='day', y='total_bill', data=tips)
plt.title('Violin Plot of Total Bill by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill ($)')
plt.show()
```

```python
## Histogram
sns.histplot(tips['total_bill'], bins=20, kde=True)
```

```python
##Kde Plot
sns.kdeplot(tips['total_bill'],fill=True)
plt.title('KDE Plot of Total Bill')
plt.xlabel('Total Bill ($)')
plt.ylabel('Density')
plt.show()
```

```python
##Pair Plot
sns.pairplot(tips, hue='day', diag_kind='kde')
plt.suptitle('Pair Plot of Tips Dataset', y=1.02)
plt.show()
```

```python
###Heatmap
corr=tips[['total_bill','tip','size']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap of Tips Dataset Correlation')
plt.show()
```
