```python
df['Total_mean']=df['Total Revenue'].fillna(df['Total Revenue'].mean())
df.head()
```

```python
df.dtypes
```

```python
df.rename(columns={'Date': 'Sales_Date'}, inplace=True)
df.head()
```

```python
df['Total Revenue']=df['Total Revenue'].astype(int)
```

```python
df.dtypes
```

```python
df['New Column'] = df['Total Revenue'].apply(lambda x: x * 2)
df.head()
```

```python
## Data Aggregation and Grouping
grouped_mean=df.groupby('Product Category')['Total Revenue'].mean()
print(grouped_mean)
```

```python
df.head()
```

```python
gropuped_sum=df.groupby(['Product Category','Product Name'])['Total Revenue'].sum()
print(gropuped_sum)
```

```python
grouped_agg=df.groupby('Product Category').agg({'Total Revenue': ['mean', 'sum', 'count']})
print(grouped_agg)
```
